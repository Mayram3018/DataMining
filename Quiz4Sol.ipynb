## Quiz 4 Solution

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
# load train.csv file from Kaggle "House Prices: Advanced Regression Techniques" dataset 
house = pd.read_csv("../../datasets/house_prices/train.csv")
```

```python
house.columns
```

```python
cols = ['LotArea', 'YearBuilt', 'MSZoning']
```

```python
X = house[cols]
X
```

```python
X = pd.get_dummies(X)
X
```

```python
y = house.loc[:,'SalePrice']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1)
lm = linear_model.LinearRegression()
model = lm.fit(X_train, y_train)
test_predictions = model.predict(X_test)
print(mean_absolute_error(y_test, test_predictions))
print(model.coef_)
print(model.intercept_)
```
